```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm, iqr, mode
```

```python
# Data
x = [7, 6, 3, 8, 12, 9, 9, 13, 13, 9, 10]
y = [1, 3, 5, 7, 9, 9, 9, 11, 13, 15, 17]
```

```python
# Mean
mean_x = np.mean(x)
mean_y = np.mean(y)
```

```python
# Median
median_x = np.median(x)
median_y = np.median(y)
```

```python
# Mode
mode_x = mode(x)
mode_y = mode(y)
```

```python
# Range
range_x = max(x) - min(x)
range_y = max(y) - min(y)
```

```python
# Interquartile Range (IQR)
iqr_x = iqr(x)
iqr_y = iqr(y)
```

```python
# Variance
variance_x = np.var(x, ddof=1)  # Sample variance
variance_y = np.var(y, ddof=1)
```

```python
# Standard Deviation
std_dev_x = np.std(x, ddof=1)  # Sample standard deviation
std_dev_y = np.std(y, ddof=1)
```

```python
# Print all calculated values
print("Statistics for x:")
print(f"Mean: {mean_x}")
print(f"Median: {median_x}")
print(f"Mode: {mode_x}")
print(f"Range: {range_x}")
print(f"IQR: {iqr_x}")
print(f"Variance: {variance_x}")
print(f"Standard Deviation: {std_dev_x}")
```

```python
print("\nStatistics for y:")
print(f"Mean: {mean_y}")
print(f"Median: {median_y}")
print(f"Mode: {mode_y}")
print(f"Range: {range_y}")
print(f"IQR: {iqr_y}")
print(f"Variance: {variance_y}")
print(f"Standard Deviation: {std_dev_y}")
```

```python
# Normal distribution plot for x
x_values = np.linspace(min(x) - 2, max(x) + 2, 1000)
normal_dist_x = norm.pdf(x_values, mean_x, std_dev_x)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_values, normal_dist_x, label='Normal Distribution (x)', color='blue')
plt.title('Normal Distribution for x')
plt.xlabel('x')
plt.ylabel('Density')
plt.axvline(mean_x, color='red', linestyle='dashed', linewidth=1, label=f'Mean ({mean_x})')
plt.axvline(mean_x + std_dev_x, color='green', linestyle='dashed', linewidth=1, label=f'+1 SD ({mean_x + std_dev_x})')
plt.axvline(mean_x - std_dev_x, color='green', linestyle='dashed', linewidth=1, label=f'-1 SD ({mean_x - std_dev_x})')
plt.legend()
plt.show()
```
